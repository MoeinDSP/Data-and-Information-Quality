## Data Quality Assessment

#### Import Libraries

```python
import pandas as pd
import os

DATA_PATH = os.path.join(os.getcwd(), 'Raw_Data')
```

#### Read Data

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'met_museum_objects.csv'))
```

#### Print Head 5

```python
df.head(5)
```

#### Dataset General Information

```python
# shape
print("Shape of the dataset:", df.shape)

# columns
print("Columns and Types in the dataset:")
for col, typ in zip(df.columns, df.dtypes):
    print("-", col, " : ", typ)
```

```python
print('Data Info')
print(df.info())
```

```python
# Number of null in each column
print('Number of nulls in each column:')
print(df.isnull().sum())

print('\n')
# Percentage of null in each column
null_percentage = (df.isnull().sum() / len(df)) * 100
null_percentage = null_percentage.sort_values(ascending=False)
print('\nColumns sorted by null percentage:')
print(null_percentage)
```

#### Unique Values

```python
print('\nUnique values in each column:')
columns_sorted = null_percentage.index.to_list()
columns_sorted.reverse()
for col in columns_sorted:
    unique_values = df[col].nunique()
    print(f"- {col}: {unique_values} unique values")
```

```python
print('\nColumns sorted by number of unique values:')
unique_values_count = {col: df[col].nunique() for col in df.columns}
sorted_unique_values = dict(sorted(unique_values_count.items(), key=lambda item: item[1], reverse=True))
for col, count in sorted_unique_values.items():
    print(f"- {col}: {count} unique values")
```

```python
print('\nColumns with less than 100 unique values:')
for col, count in sorted_unique_values.items():
    if count < 100:
        unique_vals = df[col].unique()
        print(f"- {col}: {" - ".join(map(str, unique_vals))}")
```

##### Value Counts for last part selected columns

```python
selected_columns = ['Department', 'Is Highlight', 'Is Public Domain', 'Metadata Date', 'Repository']

for col in selected_columns:
    print(f'\nValue counts for column: {col}')
    print(str(df[col].value_counts()))
    print('\n')
```

#### Duplications

```python
duplicates = df.duplicated()
print(f'Total number of duplicate rows in the dataset: {duplicates.sum()}')
print("Any duplicate rows present:", duplicates.any())
```

##### Not Null Values

```python
number_of_not_null_values = df.count().sum()
print("Total Number of not null values in the dataset:", number_of_not_null_values)
print("Number of Not Null Values for each Column:")
print(df.count())
```

##### Null Values

```python
number_of_nulls = df.isnull().sum().sum()
print("Total Number of null values in the dataset:", number_of_nulls)
print("Number of null values for each column:")
print(df.isnull().sum())
```

```python
total_number_of_cells = df.shape[0] * df.shape[1]
print("Total number of cells in the dataset:", total_number_of_cells)
```

##### Completeness Evaluation

```python
completness = (number_of_not_null_values / total_number_of_cells) * 100
print(f"Overall Completeness of the dataset: {completness:.4f}%")
```
